```python
'''STEP 12 --> fixing imbalanced data'''
#https://machinelearningmastery.com/random-oversampling-and-undersampling-for-imbalanced-classification/

#start from the original dataframe 
balancedDf = df.copy()

#get the amount of 0's and amount of 1's
zeroBalanced = balancedDf['y'].value_counts()['no'] #amount of 0
oneBalanced = balancedDf['y'].value_counts()['yes'] #amount of 1

#setting the weights for the sample
weights = {'no': oneBalanced,
           'yes': zeroBalanced}

#create a new column that assigns the weights to the corresponding response
balancedDf['weights'] = balancedDf['y'].apply(lambda x: weights[x])

balanced = balancedDf.sample(frac=1.0, replace=True, random_state=(2021-10-25), weights = 'weights')

print(len(balanced))
print(balanced['y'].value_counts()['no'])
print(balanced['y'].value_counts()['yes'])
#the amounts become our weights
#weight of no = zeroBalanced...weight of yes = oneBalanced
#we need to inverse these and give 1 the weight of 0 and vice versa
#create new column -> where y is 'no' the weight is oneBalanced value
    #else the value of 'yes' becomes zeroBalanced

```

```python
'''STEP 13 --> Using the balanced set...now redo the training steps and see what we get'''
balanced_y = balanced.iloc[0:, 20] #this is the output variable dataframe
balanced_X = balanced.iloc[0:, 0:20] #this is the rest of the factors except the output variable

#there is only one 'default yes' so we get rid of it to see how it fits
#balanced_X.loc[(balanced_X.default == 'yes'), 'default'] = 'unknown'

#from the balanced X and Y --> get balanced training/testing sets
X_bal_train, X_bal_test, y_bal_train, y_bal_test = train_test_split(balanced_X, balanced_y, test_size=0.1, random_state=(2021-10-25)) # 10% test, 90% train


#drop the duration column from the sets
X_bal_train = X_bal_train.drop(columns=['duration'])
X_bal_test = X_bal_test.drop(columns=['duration'])

#only focus on bank client data and make dummy variables
#drop all the columns after column 7
X_bal_train = X_bal_train.drop(columns=['contact','month', 'day_of_week', 'campaign','pdays','previous','poutcome','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed'])
X_bal_test = X_bal_test.drop(columns=['contact','month', 'day_of_week','campaign','pdays','previous','poutcome','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed'])

#make the dummy variables for the categorical
#categorical vars in a list 
listOfCats = ['job','marital','education','default','housing','loan']

X_bal_train = pd.get_dummies(data=X_bal_train, columns=listOfCats)
X_bal_test = pd.get_dummies(data=X_bal_test, columns=listOfCats)

#turn the bal with dummies into a dataframe
X_bal_train = pd.DataFrame(X_bal_train)
X_bal_test = pd.DataFrame(X_bal_test)

X_bal_test['default_yes'] = 0
X_bal_train.to_csv('train.csv')
X_bal_test.to_csv('test.csv')

```
