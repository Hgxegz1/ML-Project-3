```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=";")
df.head()
```

```python
y = df.iloc[0:, 20] #this is the output variable dataframe
X = df.iloc[0:, 0:20] #this is the rest of the factors except the output variable


```

```python
'''STEP 2'''
from sklearn.model_selection import train_test_split

#we use train_test_split to randomly generate sets and we also split the output value already
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=(2021-10-25)) # 20% test, 80% train

```

```python
'''STEP 3'''
#drop the 'duration' column from both train and test since it will affect the output highly
X_train = X_train.drop(columns=['duration'])
X_test = X_test.drop(columns=['duration'])

#X_train
```

```python
'''STEP 4 -> turn categorical variables between 1-10 into dummy variables'''
#categorical vars in a list 
listOfCats = ['job','marital','education','default','housing','loan','contact','month','day_of_week']
X_train = pd.get_dummies(data=X_train, columns=['job','marital','education','default','housing','loan','contact','month','day_of_week'])
X_test = pd.get_dummies(data=X_test, columns=['job','marital','education','default','housing','loan','contact','month','day_of_week'])

X_train


```
